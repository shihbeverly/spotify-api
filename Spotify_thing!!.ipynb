```python
import requests # library that helps you request things

# 🕵️ Step 2: Who are we searching for?
artist_name = "Taylor Swift"

# 📨 Step 3: Set up the search request
url = "https://api.spotify.com/v1/search"
headers = {
    "Authorization": f"Bearer {access_token}"  # Your VIP pass
}
params = {
    "q": artist_name,
    "type": "artist",
    "limit": 1  # Just give us the best match
}

# 🚀 Step 4: Send the request
response = requests.get(url, headers=headers, params=params)

# 📖 Step 5: Read and print the results
if response.status_code == 200: # parts of the response: code is if there's an error code
    data = response.json() #response.json() is like the meat of what we got back from Spotify but it's a full onion we have to peel
    artist = data['artists']['items'][0]
    print(f"Artist Name: {artist['name']}")
    print(f"Spotify ID: {artist['id']}")
    print(f"Genres: {artist['genres']}")
else:
    print("Oops! Something went wrong.")
    print("Status code:", response.status_code)
    print(response.text)
```

```python
import requests

# 🧑‍🎤 Taylor Swift's artist ID
artist_id = "06HL4z0CvFAxyc27GXpf02"

# 📍 Country to fetch top tracks for
country = "US"  # you can try "GB", "CA", etc.

# 🎯 URL to get top tracks
url = f"https://api.spotify.com/v1/artists/{artist_id}/top-tracks"

# Headers with VIP pass
headers = {
    "Authorization": f"Bearer {access_token}"
}

# Parameters
params = {
    "country": country
}

# 🚀 Send the request
response = requests.get(url, headers=headers, params=params)

# 📖 Process and print the response
if response.status_code == 200:
    data = response.json()
    print("🎧 Top Tracks:")
    for idx, track in enumerate(data['tracks'], start=1):
        print(f"{idx}. {track['album']['name']}: {track['name']} (Popularity: {track['popularity']})") # album name is nested further within the onion
else:
    print("Something went wrong:", response.status_code)
    print(response.text)
```
